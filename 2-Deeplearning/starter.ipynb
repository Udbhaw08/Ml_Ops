```python
from sklearn.model_selection import validation_curve
from hyperopt import STATUS_OK
import numpy as np
import keras
import mlflow

# Turn off automatic logging to prevent epoch-by-epoch disk writes
mlflow.tensorflow.autolog(disable=True)

def train_model(params, epochs, train_x, train_y, valid_x, valid_y, test_x, test_y):
    # Clear the Keras session to free RAM from previous models
    keras.backend.clear_session()

    mean = np.mean(train_x, axis=0)
    var = np.var(train_x, axis=0)

    model = keras.Sequential([
        keras.Input(shape=(train_x.shape[1],)),
        keras.layers.Normalization(mean=mean, variance=var),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(1)
    ])

    model.compile(
        optimizer=keras.optimizers.SGD(
            learning_rate=params['lr'], 
            momentum=params["momentum"]
        ),
        loss="mean_squared_error",
        metrics=[keras.metrics.RootMeanSquaredError()]
    )

    with mlflow.start_run(nested=True):
        model.fit(
            train_x, 
            train_y, 
            validation_data=(valid_x, valid_y),
            epochs=epochs,
            batch_size=128,  # Larger batch size to reduce iteration overhead
            verbose=0
        )

        eval_result = model.evaluate(valid_x, valid_y, batch_size=128, verbose=0)
        eval_rmse = eval_result[1]

        mlflow.log_params(params)
        mlflow.log_metric("eval_rmse", eval_rmse)
        
        # Only log the model parameters/metrics to speed up the loop.
        # Log the model artifact only after finding the best parameters.

        return {"loss": eval_rmse, "status": STATUS_OK, "model": model}
```
